```python
class Node:

    def __init__(self, data=None, next=None) -> None:
        '''
        data: data of input node (str/int)
        next: pointer to the next node
        '''
        self.data = data
        self.next = next


class LinkedList:

    def __init__(self) -> None:
        self.head = None

    def traverse(self):
        # print("Traversing the LL", end="\n")
        # HEAD points to NULL
        if self.head is None:
            print("LL is empty")
            return
        _ = self.head  # iterator
        llstr = ""  # empty string to store LL values
        while _:  # HEAD != NULL
            llstr += str(_.data) + ' --> ' if _.next else str(_.data)
            _ = _.next
        print(llstr)

    def getLength(self):
        if self.head is None:
            print("LL is empty")
            return
        count = 0
        _ = self.head
        while _:  # HEAD != NULL
            count += 1
            _ = _.next
        print("LinkedList length: ", count)
        return count

    def insertAtBeginning(self, data):
        '''
        data: data of input node (str/int)
        '''
        print("Inserting at Beginning ", end="\n")
        node = Node(data=data, next=self.head)  # NEXT points to the HEAD
        #Updating HEAD to point the current node as it's the new HEAD
        self.head = node
        # return

    def insertAtEnd(self, data):
        '''
        data: data of input node (str/int)
        '''
        print("Inserting at End ", end="\n")
        if self.head is None:  # if LL is empty
            self.insertAtBeginning(data=data)
            return
        _ = self.head  # start from HEAD to find the End
        while _.next:  # go until you find End
            _ = _.next
        _.next = Node(data=data, next=None)  # Now NEXT points to the new node

    def insertAtPosition(self, position, data):
        if position < 0 or position > self.getLength():
            #we must not insert at an invalid position
            raise Exception("Invalid position")
        elif position == 0:  # trying to insert at the start
            self.insertAtBeginning(data=data)
        elif (position == self.getLength() - 1):  # trying to insert at the end
            self.insertAtEnd(data=data)
        else:
            count = 0  # let's reach the position
            _ = self.head
            while _:
                if (count == position - 1):
                    # at this point, HEAD's NEXT points to  node at position +1
                    node = Node(data=data, next=_.next)
                    # new node's NEXT points to HEAD's NEXT
                    _.next = node
                    break
                _ = _.next
                count += 1

    # removing a Node can be handled from one function
    def removeNode(self, position):
        if position < 0 or position >= self.getLength():
            print("invalid position")
            return
        if self.head is None:
            print(" LL is empty")
            return
        print("removing node from position: ", position)
        if position == 0:
            self.head = self.head.next  #HEAD points to the NEXT node and python will do the garbage cleanup
            return
        count = 0
        _ = self.head
        while _:
            if count == position - 1:
                _.next = _.next.next
                break
            _ = _.next
            count += 1


if __name__ == "__main__":
    ll = LinkedList()
    ll.traverse()  # LL is empty
    ll.insertAtBeginning(9)
    ll.traverse()
    ll.insertAtBeginning(8)
    ll.traverse()
    ll.insertAtEnd(10)
    ll.traverse()  # 8 --> 9 --> 10
    ll.insertAtEnd(11)
    ll.traverse()
    # print(ll.getLength())
    # ll.getLength()
    ll.removeNode(position=0)
    ll.traverse()
    ll.getLength()
    ll.insertAtEnd(12)
    ll.traverse()
    ll.removeNode(position=3)
    ll.traverse()
    ll.insertAtEnd(13)
    ll.traverse()
    ll.removeNode(1)
    ll.traverse()
```
